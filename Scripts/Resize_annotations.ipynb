```python
#Rectified version
import csv

def scale_annotations(annotations_path, new_annotations_path, old_width, old_height, new_width, new_height):
    x_scale = new_width / old_width
    y_scale = new_height / old_height

    with open(annotations_path, 'r', newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=';')
        annotations = list(reader)

    with open(new_annotations_path, 'w', newline='') as new_csvfile:
        writer = csv.writer(new_csvfile, delimiter=',')

        for annotation in annotations:
            scaled_annotation = [annotation[0]]  # Keep the first element (filename or frame)

            # Scale only the x, y pairs and avoid adding extra empty strings
            for i in range(1, len(annotation) - 1, 2):
                x = annotation[i]
                y = annotation[i + 1]

                # Debug print statement to inspect values
                print(f"Original (x, y): ({x}, {y})")

                if x.strip() and y.strip():
                    try:
                        original_x = float(x)
                        original_y = float(y)
                        scaled_x = original_x * x_scale
                        scaled_y = original_y * y_scale
                        scaled_annotation.extend([scaled_x, scaled_y])

                        # Debug print statement to inspect scaled values
                        print(f"Scaled (x, y): ({scaled_x}, {scaled_y})")
                    except ValueError:
                        # If there's a problem converting x or y, just append empty strings
                        scaled_annotation.extend(['', ''])
                else:
                    # Append empty strings if x or y is missing
                    scaled_annotation.extend(['', ''])

            # Handle any remaining elements after the last x, y pair
            if len(annotation) % 2 == 0:
                scaled_annotation.append(annotation[-1])

            writer.writerow(scaled_annotation)

    print(f"Scaled annotations have been saved to: {new_annotations_path}")

# Example usage
annotations_path = r"C:\Users\hp\OneDrive - Efrei\TahaDataCorrectlyAnnotated\Canter\icelandic horse - galop 2.csv"
new_annotations_path = r"C:\Users\hp\OneDrive - Efrei\TahaDataCorrectlyAnnotated\Canter\scaled_annotations.csv"
old_width = 640
old_height = 360
new_width = 1280
new_height = 720

scale_annotations(annotations_path, new_annotations_path, old_width, old_height, new_width, new_height)
```
