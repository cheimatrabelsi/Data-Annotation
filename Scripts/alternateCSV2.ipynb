```python
#this cersion is for csv files without file name
import pandas as pd
import os

# Function to format the number into 'frame_XXXX' format
def format_frame_number(number):
    return f'{int(number):05d}'

# Path to your CSV file
csv_file_path = r"C:\Users\hp\OneDrive - Efrei\TahaDataCorrectlyAnnotated\Canter\scaled_annotations.csv"

# Read the CSV file without considering the first row as the header
df = pd.read_csv(csv_file_path, delimiter=',', header=None)

# Drop the first two rows
df = df.iloc[2:].reset_index(drop=True)

# Replace the first column values with the formatted frame numbers
df[0] = df.iloc[:, 0].apply(format_frame_number)

# Add new columns for width and height with fixed values at the end of the DataFrame
df[len(df.columns)] = 1280  # Add column for width
df[len(df.columns)] = 720   # Add column for height


# Save the updated DataFrame to a new CSV file without header
output_csv_file_path = r"C:\Users\hp\OneDrive - Efrei\TahaDataCorrectlyAnnotated\Canter\icelandic horse - galop 2Final.csv"
df.to_csv(output_csv_file_path, index=False, header=False)

print(f'Updated CSV file saved as {output_csv_file_path}')
```
