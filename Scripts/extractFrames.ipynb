```python
import cv2
import os

def extract_frames_from_video(video_path, output_dir):
    # Open the video file
    video_capture = cv2.VideoCapture(video_path)
    if not video_capture.isOpened():
        print(f"Error: Could not open video file {video_path}.")
        return

    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    frame_count = 0
    success, frame = video_capture.read()

    while success:
        # Save each frame as a JPEG image
        frame_filename = f"{output_dir}/frame_{frame_count:04}.jpg"
        cv2.imwrite(frame_filename, frame)
        frame_count += 1

        # Read the next frame
        success, frame = video_capture.read()

    video_capture.release()
    print(f"All frames extracted from {video_path}.")
    print(f"Frames saved to: {output_dir}")


# Example usage
videos_folder = r"C:\Users\hp\OneDrive - Efrei\Final Final New dataset\Trot\Videos\V29.mp4"
output_base_dir = r"C:\Users\hp\OneDrive - Efrei\Final Final New dataset\Trot\Videos\V29"

extract_frames_from_video(videos_folder, output_base_dir)
```
