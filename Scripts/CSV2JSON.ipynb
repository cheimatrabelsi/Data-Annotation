```python
import csv
import json

# Initialize the COCO data structure
coco_data = {
    "images": [],
    "annotations": [],
    "categories": [
        {"supercategory": "animal",
            "id": 1,
            "name": "horse",
            "keypoints": [
                "T1", "T2", "A8", "A7", "A6", "A5", "A3", "A1", "A7_1", "A6_1", "A5_1", "A3_1", "A1_1",
                "A10", "P9", "P7", "P5", "P3", "P1", "P9_1", "P7_1", "P5_1", "P3_1", "P1_1", "Q1", "Q2", "Q3", "Q4"
            ],
            "skeleton": []
        }
    ]
}

# Track the unique ID for each image
image_id = 0
#csv_file_path = r"C:\Users\hp\Documents\EFREI_PFE\15_05_2024\Gait Dataset\Annotated old dataset 5frames\Walk\Processed\All Gaits For Dataset - walk 1\all_gaits_for_dataset_walk_with_filenames.csv"
csv_file_path = r"C:\Users\hp\OneDrive - Efrei\TahaDataCorrectlyAnnotated\Canter\FEI - galop 1Final.csv"
# Open the CSV file and read its contents
with open(csv_file_path, newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        # Construct the file path
        file_path = f"{row[0]}"

        # Actual width and height of the image
        image_width = int(row[58])
        image_height = int(row[59])

        # Add image information
        coco_data["images"].append({
            "file_name": file_path,
            "height": image_height,
            "width": image_width,
            "id": image_id  # Assign a unique ID using the loop variable
        })

        # Process keypoint data
        keypoints = []
        num_keypoints = 0
        for i in range(2, 58, 2):  # Start from the fourth column, read every two columns, for a total of 28 keypoints
            x = row[i]
            y = row[i+1]
            if x == '' or x == ' ' or x == '0' or y == '' or y == ' ' or y == '0':
                keypoints.extend([0, 0, 0])  # Invisible keypoints
            else:
                keypoints.extend([float(x), float(y), 2])  # Visible keypoints
                num_keypoints += 1

        # Add annotation information
        coco_data["annotations"].append({
            "keypoints": keypoints,
            "image_id": image_id,
            "id": image_id,
            "num_keypoints": num_keypoints,
            "iscrowd": 0,
            "category_id": 1,
            "segmentation": [],  # Add an empty segmentation field
            "bbox": [0, 0, 1280, 720]  # Add dummy bbox coordinates
        })

        # Increment image ID
        image_id += 1
json_file_path = r"C:\Users\hp\OneDrive - Efrei\TahaDataCorrectlyAnnotated\Canter\FEI - galop 1Final.csvannotations.json"
# Write the COCO data to a JSON file
with open(json_file_path, 'w') as jsonfile:
    json.dump(coco_data, jsonfile, indent=4)
```
