```python
import json

#remove a frame frm COCO format annotation

# Specify input and output file paths
input_file_path = r"C:\Users\hp\Downloads\V33\annotations\person_keypoints_default.json"
output_file_path = r"C:\Users\hp\Downloads\V33\annotations\V33without first frame.json"

# Load the JSON file
with open(input_file_path, 'r') as f:
    data = json.load(f)

# Remove the image entry for frame_0000
data['images'] = [img for img in data['images'] if img['file_name'] != 'frame_0000.jpg']

# Remove annotations associated with the removed image
data['annotations'] = [ann for ann in data['annotations'] if ann['image_id'] != 1]

# Update image IDs
for i, img in enumerate(data['images'], start=1):
    img['id'] = i

# Update annotation image_ids
for ann in data['annotations']:
    ann['image_id'] -= 1

# Save the modified JSON
with open(output_file_path, 'w') as f:
    json.dump(data, f, indent=2)

print("Processing complete. Modified JSON saved to:", output_file_path)
```
